```python
from scipy import stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
df=pd.read_csv('Data For The Prediction.csv')
x_axis=df['rating']
y_axis=df['discount']
df['rating'].fillna(100,inplace=True)
df['discount'].fillna(100,inplace=True)
model=np.poly1d(np.polyfit(x_axis,y_axis,10))
curve=np.linspace(1,5,1000)
plt.scatter(x_axis,y_axis)
plt.plot(curve,model(curve),'g')
plt.xlabel("Rating")
plt.ylabel("Discount")
plt.title("Graph of Discount vs Rating")
plt.show()
```

```python
df=pd.read_csv('Data For The Prediction.csv')
x_axis=df['quantity']
y_axis=df['discount']
df['quantity'].fillna(100,inplace=True)
df['discount'].fillna(100,inplace=True)
model=np.poly1d(np.polyfit(x_axis,y_axis,10))
curve=np.linspace(0.5,100,1000)
plt.scatter(x_axis,y_axis)
plt.plot(curve,model(curve),'g')
plt.xlabel("Quantity")
plt.ylabel("Discount")
plt.title("Graph of Discount vs Quantity")
plt.show()
```

```python
df=pd.read_csv('Data For The Prediction.csv')
df['quantity'].fillna(100,inplace=True)
df['rating'].fillna(100,inplace=True)
df['discount'].fillna(100,inplace=True)
x_axis=df[['quantity','rating']].values
y_axis=df['discount'].values
multiple_linear_regression=linear_model.LinearRegression()
multiple_linear_regression.fit(x_axis,y_axis)
```

```python
a=int(input("Enter quantity:"))
b=float(input("Enter rating:"))
prediction=multiple_linear_regression.predict([[a,b]])
print("The Discount is:",prediction,"%")
```
